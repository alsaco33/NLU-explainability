```python
# To move to metric extraction
import re

def feature_agreement(explanation, tokenized_selected_text, k):
  """
  Fraction of top k tokens in the predicted saliency map that are common with the ground truth explanation

  args:
    explanations: Predicted explanations
    selected_text: Ground truth annotated explanation
    k: Cutoff of number of tokens to keep 
  
  returns:
    feat_agreement: Percentage of coincidences

  """

  #Define value at which to cut to keep k records
  cutoff = sorted([explanation[i][1] for i in range(len(explanation))])[-k:][0]
  
  # Select tokens in explanation with higher weight than cutoff
  top_explanation = [c[0] for c in explanation if c[1]>=cutoff]
  top_explanation_clean = [token.replace('_', '').replace('Ġ', '').replace('#', '').replace('@', '').lower() for token in top_explanation]

  # Compute overlap size
  tokenized_selected_text_clean = [token.replace('_', '').replace('Ġ', '').replace('#', '').replace('@', '').lower() for token in tokenized_selected_text]
  overlap = [token for token in tokenized_selected_text_clean if token in top_explanation_clean]

  # Normalize by length of top explanation
  fa      = len(overlap)/len(top_explanation_clean)
  return fa

def iou(explanation, tokenized_selected_text, k):
  """
  For two spans of text (human explanation and model explanation), it is defined as the size of the overlap of the tokens of the two 
  spans divided by the size of their union.
  """
  
  # Define value at which to cut to keep k records
  cutoff = sorted([explanation[i][1] for i in range(len(explanation))])[-k:][0]

  # Select tokens in explanation with higher weight than cutoff
  top_explanation = [c[0] for c in explanation if c[1]>=cutoff]
  top_explanation_clean = [token.replace('_', '').replace('Ġ', '').replace('#', '').replace('@', '').lower() for token in top_explanation]

  # Compute overlap size
  tokenized_selected_text_clean = [token.replace('_', '').replace('Ġ', '').replace('#', '').replace('@', '').lower() for token in tokenized_selected_text]
  overlap = [token for token in tokenized_selected_text_clean if token in top_explanation_clean]

  # Compute union size
  union = len(top_explanation_clean) + len(tokenized_selected_text) - len(overlap)

  # Compute IoU
  iou = len(overlap) / union

  return iou
```
